```python
 movies4['time_diffrence'] = movies4["time_diffrence"].apply(lambda x: int(x.days))
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
sns.set_style('whitegrid')
movies5 = movies4[(movies4['time_diffrence'] > 1) & (movies4['ww_rev'] > 50000000)]
movies5['profit_velocity'] = movies4['ww_rev']/movies4['time_diffrence']
movies5['profit_velocity'].describe()
# movies5['time_diffrence'].describe()


profit_velocity = sns.scatterplot(x = movies5['time_diffrence'], y = movies5['ww_rev'], hue=movies5['time_diffrence'])
profit_velocity.set(xlabel = 'Time Since Released (days)', ylabel = 'Worldwide Revenue')
plt.title('Profit Velocity')
plt.xlim(300, 20000)
plt.ylim(50000000, 475000000)
profit_velocity.set_yticklabels(['$0M', '$100M', '$150M', '$200M', '$250M', '$300M', '$350M', '$400M', '$450M'])
# plt.xlim(10000000, 800000000)
```

```python
movies5['profit_velocity'].describe()
```

```python
top10 = movies5[['title', 'profit_velocity']].sort_values('profit_velocity', ascending=False)[0:10]
names = top10['title'].to_list()
# names = movies5[['title', 'profit_velocity']].sort_values('profit_velocity', ascending=False)[0:10]['titles'].to_list()
```

```python
top10_velocity = sns.barplot(x=names, y=movies5[['title', 'profit_velocity']].sort_values('profit_velocity', ascending=False)['profit_velocity'][0:10])
top10_velocity.set_xticklabels(
    top10_velocity.get_xticklabels(),
    rotation = 45, horizontalalignment = 'right')
plt.title('Top 10 Fastest Profiting Movies')
top10_velocity.set(xlabel='Movies', ylabel='Profit Velocity ($/day)')
```
