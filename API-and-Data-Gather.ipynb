```python
def retrieve_money(soup):
    warning = soup.find('div', class_="alert alert-warning")
    movie_container = warning.nextSibling.nextSibling
    money = [h3.find('a').attrs['a-text-right mojo-field-type-money'] for h3 in movie_container.findAll('h3')]
    return money
```
