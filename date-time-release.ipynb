```python
release_dates = []
for release in movies_df['release_date']:
    release_dates.append(release)
release_dates
```

```python
string_release = []
for i in release_dates:
    string_release.append(str(i))
type(string_release[0])
```

```python
for i in string_release:
    string_release.remove('nan')
```

```python
dt_relase = []
for i in string_release:
     dt_relase.append(dt.datetime.strptime(i, '%Y-%m-%d'))
dt_relase
```

```python
len(dt_relase)
```

```python
release_date_df = pd.DataFrame({'movie_title': movies_df['title'], 'release_date': release_dates})
```

```python
release_date_df.set_index('movie_title', inplace = True)
```

```python
release_date_df.dropna(inplace = True)
```

```python
release_date_df['release_date_object'] = dt_relase
```

```python
release_date_df.drop(columns = ['release_date'], axis = 1, inplace = True)
```

```python
release_date_df['current_date'] = dt.datetime(2020, 3, 6)
```

```python
release_date_df
```

```python
release_date_df['time_diffrence'] = (release_date_df['current_date']- release_date_df['release_date_object'])
```
